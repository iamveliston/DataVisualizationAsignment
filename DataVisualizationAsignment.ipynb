# Plotly Assignment

1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', title="3D Scatter Plot",
                    labels={'X': 'X-Axis', 'Y': 'Y-Axis', 'Z': 'Z-Axis'})

fig.show()
```

2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

df = pd.DataFrame(data)

fig = px.violin(df, x='Grade', y='Score', box=True, points="all",
                title="Violin Plot of Scores by Grade", labels={'Grade': 'Grades', 'Score': 'Scores'})

fig.show()

```

3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

fig = px.density_heatmap(df, x='Day', y='Month', z='Sales',
                         title="Sales Variation Heatmap", labels={'Day': 'Day of the Month', 'Month': 'Month'})

fig.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function:

```python
import numpy as np
import pandas as pd
import plotly.express as px

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Z',
                    title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))")

fig.show()
```

5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x-axis), and bubble size proportional to the population.

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

fig = px.scatter(df, x='GDP', y='Population', size='Population',
                 hover_name='Country', title="Bubble Chart of Population vs GDP",
                 labels={'GDP': 'GDP (in billions)', 'Population': 'Population (in millions)'})

fig.show()
```

# Bokeh Assignment

1. Create a Bokeh plot displaying a sine wave. Set x values from 0 to 10 and y values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x, y, legend_label="Sine", line_width=2)

show(p)
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the sizes and colors columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
import numpy as np
import pandas as pd

output_notebook()

np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 80, size=100)
colors = np.random.randint(10, 80, size=100)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

color_mapper = LinearColorMapper(palette="Viridis256", low=min(colors), high=max(colors))

p = figure(title="Random Scatter Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

p.scatter(x='x', y='y', size='sizes', color={'field': 'colors', 'transform': color_mapper},
          alpha=0.6, source=source)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0), title="Color Scale")
p.add_layout(color_bar, 'right')

show(p)

```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

data = np.random.randn(1000)
hist, edges = np.histogram(data, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="white")

show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.io import output_notebook
import numpy as np

output_notebook()

data = np.random.rand(10, 10)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')

color_mapper = LinearColorMapper(palette="Viridis256", low=data.min(), high=data.max())
p.image(image=[data], x=0, y=0, dw=10, dh=10, color_mapper=color_mapper)

show(p)
```
